python's Matplotlib.pyplot module is used for plotting. 

```
import matplotlib.pyplot as plt
```

```
# Creating Sample data set using numpy's array method.
import numpy as np

x = np.array(range(-10,11))
y = x ** 2
```

```
# plotting sample data set generated above using simple plot method from matplotlib.pyplot

# Setting title, x and y labels of the plot
plt.title("Parabolla")
plt.xlabel("x values")
plt.ylabel("y = x ** 2")

# Plotting the graph
plt.plot(x,y)
plt.show()
```

```
# more plots on the same layout using subplots

# plt.subplot(no_of_rows, no_of_columns, index_of_graph)
plt.subplot(1,2,2)
plt.plot(x,y)

plt.subplot(1,2,1)
plt.plot(range(21),x)
plt.show()
```

```
# plotting the similar things using object oriented method

# Figure is the top level container for all plots. Creating instance of figure
fig = plt.figure()

# adding axes to the invisible container
# passing list of elements to add_axes method, left, right, width, hight of the axes
axes = fig.add_axes([0.5,0.1,0.5,0.5])
axes1 = fig.add_axes([0.1,1,0.5,0.5])
axes.plot(x,y)
axes.set_title("Parabola wrt y-axis Fig1")
axes.set_xlabel("x2")
axes.set_ylabel("x")

axes1.plot(y,x)
axes1.set_title("Parabola wrt X-axis Fig2")
axes1.set_xlabel("x")
axes1.set_ylabel("x2")
plt.show()
```

```
# subplots using fig, axes
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title("y = x2")

axes[1].plot(y, x)
axes[1].set_title("x = y2")

plt.show()
```

```
# Figure size in figure and subplots

# figsize =(wide in inches, height in inches)
fig = plt.figure(figsize=(4,4))

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
plt.show()
```

```
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))

axes[0].plot(x,y, color='red',lw=2, ls='--')
axes[1].plot(y,x, color='green', linewidth=4)
plt.show()
```

```
# saving a figure as a picture

fig.savefig("Parabolas.jpg")
```

```
# Adding Legends, marker

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,1,1])

axes.plot(x, y, label='x vs x2',lw=2,ls='--', marker='o')
axes.plot(x, x, label='x vs x', lw=2, marker='*')

axes.legend()
plt.show()
```
